### 34) write a function to get the country sales, profit dynamically, if we pass any country, that 
### country sales, profit should be displayed? 

```python
def getCountrySalesAndProft(country):
    country_sales = tdata.query('Country==@country')
    
    if country_sales.empty:
        return "NO DATA"
    return country_sales[['Sales','Profit']].sum()
```

```python
getCountrySalesAndProft("Italy")
```

### 35) write a function to get the Category orderquantity dynamically, if we pass any Category, 
### that Category orderqunatity should be displayed.? 

```python
def getCategoryOrderQty(category):
    categoryQty = tdata.query('Category==@category')
    
    if categoryQty.empty:
        return "NO DATA"
    return categoryQty[['OrderQuantity']].sum()
```

### 36) write a function to get the City sales and profit dynamically, if we pass any city, that City 
### sales and profit should be displayed? 

```python
def getCitySalesProfit(city):
    citySalesProfit = tdata.query('City==@city')
    
    if citySalesProfit.empty:
        return "NO DATA"
    return citySalesProfit[['Sales','Profit']].sum()
```

```python
getCitySalesProfit("Stockholm")
```
